```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
from pyspark.sql.types import *
    
# Create the schema for the table
orderSchema = StructType([
    StructField("MatchUp", StringType()),
    StructField("IN", StringType()),
    StructField("OUT", StringType())
    ])
    
# Import all files from bronze folder of lakehouse
df = spark.read.format("csv").option("header", "true").option("delimiter", ";").schema(orderSchema).load("abfss://MTG_DEV@onelake.dfs.fabric.microsoft.com/Decks.Lakehouse/Files/SB_Guide/*.csv")
    
# Display the first 10 rows of the dataframe to preview your data
display(df.head(16))
```
